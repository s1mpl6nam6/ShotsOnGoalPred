```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("shots_per_match.csv")
```

```python
num_teams= len(df["away_team_name"].unique())
num_rows = math.ceil(num_teams/5)

fig, axes = plt.subplots(5, num_rows, figsize=(num_teams, num_teams))

for index, team in enumerate(df["away_team_name"].unique()):
    temp_away = df[(df["away_team_name"] == team)]
    temp_home = df[(df["home_team_name"] == team)]
    combined = pd.concat([
            temp_away["Away_shots_on_target"],
            temp_home["Home_shots_on_target"]
        ]).dropna()

    axes[index//num_rows][index%num_rows].hist(combined, bins = 7, edgecolor="black")
    axes[index//num_rows][index%num_rows].set_title(team)
    
plt.tight_layout()
plt.show()
```
